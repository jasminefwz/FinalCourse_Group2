```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('Impact_of_Mobile_Phone_on_Students_Health.csv')

# Display the first few rows of the dataset
data.head()
```

## Section 2: Handle Missing Values

```python
# Handle missing values (example: fill with mode or median)
for column in data.columns:
    if data[column].dtype == 'object':
        data[column].fillna(data[column].mode()[0], inplace=True)
    else:
        data[column].fillna(data[column].median(), inplace=True)

# Check for missing values to ensure none remain
missing_values = data.isnull().sum()
print("Missing values in each column:\n", missing_values)
```

There are no missing values ​​in this dataset. All columns have complete data for each entry.

## Section 3: Convert Age Ranges to Numerical Values

```python
# Function to convert age ranges to numerical values (midpoints)
def convert_age_range(age_range):
    if pd.isnull(age_range):
        return np.nan
    age_min, age_max = map(int, age_range.split('-'))
    return (age_min + age_max) / 2

# Convert Age column to numerical values
data['Age'] = data['Age'].apply(convert_age_range)

# Display the first few rows to confirm the Age conversion
data.head()
```

## Section 5: Generate Descriptive Statistics

```python
# Generate descriptive statistics for numerical columns
numerical_cols = ['Age', 'Daily usages', 'Academic Performance']
numerical_stats = data[numerical_cols].describe()
print("Numerical Columns Descriptive Statistics:\n", numerical_stats)

# Generate descriptive statistics for categorical columns
categorical_cols = data.select_dtypes(include=['object']).columns
categorical_stats = data[categorical_cols].describe()
print("\nCategorical Columns Descriptive Statistics:\n", categorical_stats)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Contoh DataFrame
data = pd.DataFrame({
    'Health rating': ['Always', 'Sometimes', 'Often', 'Rarely', 'Never', 'Sometimes', 'Often', 'Always'],
    'Symptom frequency': ['Never', 'Rarely', 'Sometimes', 'Often', 'Always', 'Never', 'Rarely', 'Sometimes'],
    'Daily usages': ['Sometimes', 'Often', 'Always', 'Never', 'Rarely', 'Sometimes', 'Often', 'Always'],
    'Performance impact': ['Rarely', 'Sometimes', 'Often', 'Always', 'Never', 'Sometimes', 'Often', 'Always']
})

# Print unique values
print("\nUnique values in 'Health rating':")
print(data['Health rating'].unique())

print("\nUnique values in 'Symptom frequency':")
print(data['Symptom frequency'].unique())

print("\nUnique values in 'Daily usages':")
print(data['Daily usages'].unique())

print("\nUnique values in 'Performance impact':")
print(data['Performance impact'].unique())

# Define mappings for categorical data
categorical_mappings = {
    'Never': 0,
    'Rarely': 1,
    'Sometimes': 2,
    'Often': 3,
    'Always': 4
}

# Apply the mappings to the appropriate columns
for column in ['Health rating', 'Symptom frequency', 'Daily usages', 'Performance impact']:
    if data[column].dtype == 'object':
        data[column] = data[column].map(categorical_mappings).fillna(data[column])

# Convert columns to numeric (in case some values were not in the mapping and remain non-numeric)
for column in ['Health rating', 'Symptom frequency', 'Daily usages', 'Performance impact']:
    data[column] = pd.to_numeric(data[column], errors='coerce')

# Drop rows with NaN values that result from conversion
data = data.dropna(subset=['Health rating', 'Symptom frequency', 'Daily usages', 'Performance impact'])

# Print the dataset after conversion
print("\nDataset after conversion:")
print(data[['Health rating', 'Symptom frequency', 'Daily usages', 'Performance impact']].head())

# Calculate the correlation matrix for the selected columns
correlation_matrix = data[['Health rating', 'Symptom frequency', 'Daily usages', 'Performance impact']].corr()

# Display the correlation matrix
print("\nCorrelation Matrix:")
print(correlation_matrix)

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix between Health and Academic Performance Variables')
plt.show()
```

## Conlusion
Overall, good health is associated with better work performance. People with better health tend to have better performance, while the frequency of symptoms does not have a significant impact on performance. Apart from that, daily activities related to health can also contribute to improving work performance.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from factor_analyzer import FactorAnalyzer

# Example dataset (replace with your actual dataset)
data = {
    'Sleep quality': [5, 4, 3, 2, 4],
    'Stress level': [2, 3, 4, 1, 3],
    'Social support': [4, 5, 3, 2, 5],
    'Physical activity': [3, 2, 4, 5, 3],
    'Mood': [5, 4, 5, 3, 4],
    'Health perception': [4, 3, 4, 2, 3]
}

# Creating a DataFrame from the data
df = pd.DataFrame(data)

# Check for multicollinearity
correlation_matrix = df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()

# Initialize FactorAnalyzer with number of factors
fa = FactorAnalyzer(n_factors=2, rotation='varimax')

try:
    # Fit FactorAnalyzer to data
    fa.fit(df)

    # Get factor loadings
    factor_loadings = fa.loadings_

    # Create a DataFrame for factor loadings
    factor_loadings_df = pd.DataFrame(factor_loadings, index=df.columns, columns=['Factor 1', 'Factor 2'])

    # Print factor loadings
    print("\nFactor Loadings:")
    print(factor_loadings_df)

    # Visualize factor loadings using a heatmap
    plt.figure(figsize=(10, 6))
    sns.heatmap(factor_loadings_df, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
    plt.title('Factor Loadings')
    plt.show()

except Exception as e:
    print(f"Error: {str(e)}")
```

## Conclusion
Social support, sleep quality, and stress level are the key predictors for supporting student well-being. Enhancing social support networks and improving sleep quality can significantly boost students' mood and health perception. Managing stress levels is also crucial for overall well-being. While physical activity shows negative correlations with mood and health perception in this matrix, it is generally beneficial for health, and the negative correlations may indicate areas where balance and integration with other factors are needed.
